```python
import pandas as pd
data=pd.read_csv("anime.csv")
df=pd.DataFrame(data)

df

```

```python
def extract_episode(text):
    i=0
    while(i<len(text)):
        data=""
        if(text[i]=="("):
            i+=1
            while(i<len(text) and text[i]!=")"):
                data+=text[i]
                i+=1
            return data
        i+=1
    return None
        
                
            
```

```python
df["episodes"]=df["Title"].apply(extract_episode)
df
```

```python
def remove_eps(text):
    data=text.strip("eps")
    return data
```

Extracting the Duration



```python
def extract_time(text):
    i=0
    data=""
    while(i<len(text)):
        if(text[i]==")"):
            i+=1
            while(i<len(text) and text[i]!=','):
                data+=text[i]
                i+=1
            return data[:-1]
        i+=1
    return None
    
```

```python
df["Duration"]=df["Title"].apply(extract_time)


df
```
